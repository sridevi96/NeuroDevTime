## <i>1. Load the processed data

```r
#load metadata for processed HNOCA dataset
HNOCA_metadata<-read.csv(file = "processed_data/HNOCA_pseudobulk_metadata_final.csv",header = TRUE)
final_common_genes<-readRDS("models/commongenes.rds")#list of common genes
```

## <i> 2. Generate rank normalized raw counts </i>
> <b> Here we will use the raw count matrix, i.e. sparse_data, for rank normalizing the expression values

```r
#Rename the columns to gene names to genes of interest and rank normalize
colnames(sparse_data)<-genedata$feature_name
dim(sparse_data) #columns are genes and rows are cells
```

```r
which(genedata$ensembl_id=='ENSG00000185010') #manually fixing one unmapped gene
```

```r
colnames(sparse_data)[15390]<-'F8' #one unmapped gene
```

```r
#restrict the matrix to the final common genes used in model training
HNOCA_counts<-t(sparse_data[,intersect(final_common_genes,colnames(sparse_data))])
dim(HNOCA_counts)
length(final_common_genes)
```

> ### Load rank normalized model

```r
ranknorm_model<-readRDS("models/ranknormalized_celltypeagnostic_model.rds")
```

```r
#check coefficients
coefs<-as.data.frame(coef(ranknorm_model$finalModel,ranknorm_model$finalModel$lambdaOpt))%>%filter(s1!=0)
coefs$gene<-rownames(coefs)
coefs
```

### Check whether model genes are present in the dataset
> Note: the predict function expects a matrix that has the same genes as what was used in model training. Some of these genes may be missing in the test dataset (in this case 10944, vs expected 10957 genes), but what matters is whether all the genes used by the model are present in the dataset. We will check this below, and zero-pad the rest, since they won't be used for age prediction anyway.

```r
length(intersect(coefs$gene[-1],rownames(HNOCA_counts))) #all model genes are found! 
#in the training data, they were normalized with respect to the final_common_genes, so we zero pad the missing ones

zero_padding<-matrix(0,nrow=length(setdiff(final_common_genes,rownames(HNOCA_counts))),ncol=ncol(HNOCA_counts))
rownames(zero_padding)<-setdiff(final_common_genes,rownames(HNOCA_counts))
HNOCA_counts<-rbind(HNOCA_counts,zero_padding)
dim(HNOCA_counts)
```

```r
ranknormalize<-function(mat){ #rows are genes and columns are cells
    ranks<-apply(mat,2,rank,ties.method ="average") #get the rank of each gene and replace 
    ranks<-ranks/nrow(mat)#divide by max rank or number of genes
    return(ranks)
}
```

```r
HNOCA_ranknorm<-ranknormalize(HNOCA_counts)#generate rank normalized counts matrix
```

```r
dim(HNOCA_ranknorm)
```

---

## <i> 3. Predict ages in HNOCA organoid atlas

```r
HNOCA_predicted_ages<-predict(ranknorm_model,t(HNOCA_ranknorm)) 
```

```r
datatoplot<-cbind(HNOCA_metadata,HNOCA_predicted_ages)
```

```r
means=datatoplot%>%group_by(organoid_age_days,publication)%>%summarize(avg=mean(HNOCA_predicted_ages),
                                                              sem=sd(HNOCA_predicted_ages)/sqrt(length(HNOCA_predicted_ages)))
```

```r
HNOCA_plot<-ggplot(datatoplot,aes(x=round(organoid_age_days/30,2),y=HNOCA_predicted_ages))+geom_point(alpha=0.2,aes(col=publication))+
geom_point(data=means,aes(x=round(organoid_age_days/30,2),y=avg,col=publication),shape='square',size=4)+
geom_smooth(method='loess')+xlab('organoid_age_months')+
stat_cor()+ggtitle("Overall predictions in HNOCA dataset")+stat_cor()
#dev.off()
HNOCA_plot
```

```r
ggplot(datatoplot,aes(x=round(organoid_age_days/30,2),y=HNOCA_predicted_ages))+geom_point(alpha=0.5,aes(col=HNOCA_predicted_ages))+
scale_color_viridis_c()+
stat_cor()+
xlab("organoid_age_months")+
facet_wrap(~annot_level_1)+ggtitle("Age predictions by cell type")

```
