```r
#2 measures of cell type specific model on self: overall performance
get_off_diagonal_mean <- function(matrix) {
  # Create a diagonal mask
  diag_mask <- row(matrix) == col(matrix)
  matrix[diag_mask] <- NA
  # Calculate and return the column means: for each model on all other datasets
  colMeans(matrix, na.rm = TRUE)
}

get_off_diagonal_median <- function(matrix) {
  # Create a diagonal mask
  diag_mask <- row(matrix) == col(matrix)
  matrix[diag_mask] <- NA
  # Calculate and return the column means: for each model on all other datasets
  apply(matrix,2,median,na.rm = TRUE)
}


#get_off_diagonal_mean(performance_celltypemodels_revised)#THIS IS CORRECT
#get_off_diagonal_median(performance_celltypemodels_revised)#THIS IS CORRECT
```

```r
assigncelltype_hypothalamus<-function(original){
  if(original %in% c("Astrocytes")){newname<-"Astrocyte"}
  else if(original %in% c("Neuronal")){newname<-"Neuron"}
  else if(original  %in% c("IntProgen-1","IntProgen-2","IntProgen-3","qRGC","cRGC","tRGC")){newname<-"Progenitor"}
  else if(original %in% c("Ependymal","Tanycytes")){newname<-"Other"}
  else if(original %in% c("Microglia")){newname<-"Immune"}
  else if(original %in% c("Oligodendrocytes [Intermediate]","Oligodendrocytes [Immature]","Oligodendrocytes [Mature]","Oligodendrocytes [Dividing]")){newname<-"Oligo"}
  else if(original %in% c("Endothelial","Pericytes","Blood","VLMC","vSMC")){newname<-"Vascular"}
  
  return(newname)
}
```

```r
predict_celltypespecific<-function(genexpdata,metadata,models){
    
    split_celltypes<-split(1:nrow(metadata),metadata$cellclass)
    result<-lapply(1:length(split_celltypes),function(i){
        celltype=names(split_celltypes)[i]
        cols=split_celltypes[[i]]
        data=genexpdata[,cols]
        if(celltype=="Neuron"){
            modeltouse1=models[["Exc_Neuron"]]
            modeltouse2=models[["Inh_Neuron"]]
            pred1= predict(modeltouse1,t(data))
            pred2= predict(modeltouse2,t(data))
            predictions=data.frame(pred1,pred2,
                              obs=metadata$obs_age[cols],
                              cellclass=celltype,
                              orig_cellclass=metadata$celltype[cols],
                              rowindex=cols,
                              study=metadata$study[cols]
                       )
        }
       else{
        modeltouse=models[[celltype]]
        pred=predict(modeltouse,t(data))
        predictions=data.frame(pred1=pred,pred2=pred,
                              obs=metadata$obs_age[cols],
                              cellclass=celltype,
                              orig_cellclass=metadata$celltype[cols],
                              rowindex=cols,
                             study=metadata$study[cols])
       }
    return(predictions)
    })%>%do.call(rbind,.) #get predictions for each celltype
    return(result)
}
```
